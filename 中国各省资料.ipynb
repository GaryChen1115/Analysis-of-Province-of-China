## 欄位的處理

```python
df1.head()
```

```python
#建新的欄位"人口密度"，降冪排序，取前五列
df1['人口密度']=df1['人口(千人)']*1000/df1['Area(km2)']
df1.sort_values('人口密度', ascending=False).head()
```

以上結論：全中國人口密度最高的行政區為上海市，每平方公里達2892人

```python
#建新的欄位"人口密度"，降冪排序，取前五列
df1['人口密度']=df1['人口(千人)']*1000/df1['Area(km2)']
df1.sort_values('人口密度', ascending=False).tail()
```

以上結論：全中國人口密度最低的行政區為西藏自治區，每平方公里僅2.7人

```python
#建新的欄位"人均GDP(美元)"，2019年7月3日US$1 = RMB$6.88，降冪排序，取前五列
df1['人均GDP(美元)']=(df1['GDP(CNY 十亿)']*1000000000)/(df1['人口(千人)']*1000*6.88)
df1.sort_values('人均GDP(美元)', ascending = False).head()
```

以上結論：北京市人均GDP為全中國最高，為US$18,740/人

```python
#建新的欄位"人均GDP(美元)"，2019年7月3日US$1 = RMB$6.88，降冪排序，取前五列
df1['人均GDP(美元)']=(df1['GDP(CNY 十亿)']*1000000000)/(df1['人口(千人)']*1000*6.88)
df1.sort_values('人均GDP(美元)', ascending = False).tail()
```

以上結論：甘肅省人均GDP為全中國最低，為US$4,285/人

```python
#先複製資料表
#新的'rain'欄位為原先'rain'欄位應用/套用一個轉換所得來的
#其中lambda x為暫時的虛擬函式，它是由原先有分號的值，透過'replace'轉換成沒有分號的值所得來的
df2 = df1
df2['rain']=df2['rain'].apply(lambda x: int(x.replace(',','')))
df2.head()
```

### 對欄位分級

```python
#複製一份表格
df3 = df1
df3.head()
```

```python
#對人口密度分四個等級：極擁擠(500人/km2以上)，擁擠(200人/km2以上)，普通(50至200人/km2)，稀疏(50人/km2以下)
df3['人口密度等級']=df3['人口密度'].apply(lambda x: '極擁擠' if (x>500) else('擁擠' if (x>200) else('普通' if (x>50) else'稀疏')))
df3
```

```python
#計算各種人口密度等級的行政區分別有多少
df3a = df3
df3a = df3.groupby('人口密度等級').count()
df3a
```

```python
#用圓餅圖呈現各種人口密度等級的占比
df3a['地区'].sort_values().plot(kind='pie')
```

以上結論：單單極擁擠(500人/km2以上)與擁擠(200人/km2以上)的省份就佔了全中國一半以上
